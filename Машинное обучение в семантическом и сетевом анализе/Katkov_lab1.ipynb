```python
import networkx as nx
import pandas as pd
import matplotlib.pyplot as plt


data = pd.read_csv("sample_data/genome_scores.csv")
data = data[::1000]
G = nx.Graph()
G.add_nodes_from(data['tagId'].unique())
G.add_nodes_from(data['movieId'].unique())
for row in data.itertuples():
    G.add_edge(row.tagId, row.movieId, weight=row.relevance)
```

```python
pos = nx.spring_layout(G)
nx.draw_networkx_nodes(G, pos, node_size=15, cmap=plt.get_cmap('viridis'))
nx.draw_networkx_edges(G, pos, alpha=0.2)

plt.show()
```
