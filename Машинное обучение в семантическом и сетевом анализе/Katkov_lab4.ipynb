```python
import networkx as nx
import matplotlib.pyplot as plt

def add_piersinski_carpet_node(G, level, parent, base_x, base_y, size):
    if level == 0:
        return
    else:
        step = size / 3
        for i in range(3):
            for j in range(3):
                if i != 1 or j != 1:
                    x = base_x + i * step
                    y = base_y + j * step
                    new_node = (x, y)
                    G.add_node(new_node)
                    G.add_edge(parent, new_node)
                    add_piersinski_carpet_node(G, level - 1, new_node, x - step, y - step, step)
```

```python
G = nx.Graph()
initial_node = (0, 0)
G.add_node(initial_node)
add_piersinski_carpet_node(G, 4, initial_node, -1, -1, 2)
pos = {node: node for node in G.nodes()}
nx.draw(G, pos, node_size=0, edge_color='b')
plt.show()
```

```python
avg_degree = sum(dict(G.degree()).values()) / len(G)
print("Средняя степень вершины:", avg_degree)
degrees = [G.degree(node) for node in G.nodes()]

plt.hist(degrees, bins='auto', rwidth=0.85)
plt.title('Распределение степеней вершин')
plt.xlabel('Степень вершины')
plt.ylabel('Частота')
plt.show()
```
